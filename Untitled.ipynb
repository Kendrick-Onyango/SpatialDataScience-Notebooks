```r
space_id <- c("geoid10")
time_label <- c("Year")
time_id <- c("2010", "2012")
timevars <- c("FALLRT","LInc","FRisk","BlkMaj","WhtMaj","HispMaj")
timeconst <- ("TractYear")
tperiods <- length(time_id)
```

```r
st_stack<- function(df,space_id,time_label,time_id,timevars,timeconst){
    dfspt <- data.frame()
    with (df, {
        for (i in time_id){
            varn<- paste(timevars,i,sep="")
            subvars<- c(space_id, varn, timeconst)
            subd <- df[, subvars]
            colnames(subd)<- c(space_id,timevars,timeconst)
            subd<- cbind(as.numeric(i),subd)
            colnames(subd)[1] <- time_label
            dfspt <<-rbind(dfspt,subd)
            }      
        dfspt <<- dfspt[,c(space_id,setdiff(names(dfspt),space_id))]
    })
    return(dfspt)
}
```

```r
dftest <- st_stack(df=chiFR,space_id=space_id,time_label=time_label,time_id=time_id,timevars=timevars,timeconst=timeconst)
```
