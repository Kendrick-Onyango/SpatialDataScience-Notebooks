```r
fm=mrate~legal+state+year # No trends, No weights

R1.DD1<- plm(fm, data=index1, model="pooling") # All deaths
R1.DD2<- plm(fm, data=index2, model="pooling") # Motor vehicle accidents
R1.DD3<- plm(fm, data=index3, model="pooling") # Suicide
R1.DD6<- plm(fm, data=index6, model="pooling") # Internal
```

```r
### CONDUCT TESTS ON FIRST RESULT:

#Breusch-Pagan LM test  confirms random effects model is needed (FAIL)
plmtest(R1.DD1, type=c("bp")) 
```

```r
#Test for x-sectional dependence
pcdtest(R1.DD1, test = c("lm")) #Breusch-Pagan LM test of independence (FAIL)
pcdtest(R1.DD1, test = c("cd")) # Pesaran CD test (PASS)
```

```r
#Test for serial correlation (FAIL)
pbgtest(R1.DD1)
```

```r
# Test for heteroskedasticity (FAIL)
bptest(R1.DD1, studentize=F)
```

```r
# We have heteroskedasticity; use robust COV matrix to account
coeftest(R1.DD1, vcovHC(R1.DD1, type = "HC1")) # Heteroskedasticity consistent coefficients, type 3
t(sapply(c("HC0", "HC1", "HC2", "HC3", "HC4"), function(x) sqrt(diag(vcovHC(R1.DD1, type = x)))))
```

```r
# We have heteroskedasticity; use robust COV matrix to account
coeftest(R1.DD2, vcovHC(R1.DD2, type = "HC1")) # Heteroskedasticity consistent coefficients, type 3
t(sapply(c("HC0", "HC1", "HC2", "HC3", "HC4"), function(x) sqrt(diag(vcovHC(R1.DD1, type = x)))))
```

```r
# FROM STUDY: All deaths = 10.80 (4.59)
# OUR RESULTS: All deaths = 10.80 (4.55)
coeftest(R1.DD1, vcov=vcovHC(R1.DD1,type="HC1",method="arellano",cluster="group"))
```

```r
# FROM STUDY: MVA = 7.59 (2.50)
# OUR RESULTS: MVA = 7.59 (2.47)
coeftest(R1.DD2, vcov=vcovHC(R1.DD2,type="HC1",method="arellano",cluster="group"))
```

```r
# FROM STUDY: Suicide = 0.59 (.59)
# OUR RESULTS: MVA = 0.59 (.58)
coeftest(R1.DD3, vcov=vcovHC(R1.DD3,type="HC1",method="arellano",cluster="group"))

```

```r
# FROM STUDY: All internal causes = 1.33 (1.59)
# OUR RESULTS: All internal causes = 1.33 (1.57)
coeftest(R1.DD6, vcov=vcovHC(R1.DD6,type="HC1",method="arellano",cluster="group"))
```

```r
# We have heteroskedasticity; use robust COV matrix to account
coeftest(R2.DD1, vcovHC(R2.DD1, type = "HC1")) # Heteroskedasticity consistent coefficients, type 3
t(sapply(c("HC0", "HC1", "HC2", "HC3", "HC4"), function(x) sqrt(diag(vcovHC(R1.DD1, type = x)))))
```

```r
##################################
##################################
##################################
# TEST FOR SPATIAL AUTOCORRELATION
##################################

#Load Shapes
states<-readOGR(".","MLDA_states",stringsAsFactors=FALSE)
class(states)
statesD<-states@data
```

```r
attach(statesD)
length(statesD)

RID

statesD$STATEFP
class(statesD$STATEFP)

temp<- as
statesD$RID <- as.numeric(statesD$STATEFP)
head(statesD)
```

```r
##################################
## Create big W matrix for panel analysis
##################################

W_state_q <- poly2nb(states,queen=FALSE)

# https://cran.r-project.org/web/packages/spdep/vignettes/nb.pdf
W_state_k4 <- read.gwt2nb("MLDA_states_k4.gwt",region.id = RID)

length(W_state_k4)

#convert weight to weight-list object
wr<- nb2listw(W_state_k4)

#convert to sparse matrix
w1<- listw2dgCMatrix(wr)
```

```r
#create time periods, validate
time_id<- c("1970","1971","1972","1973","1974","1975","1976","1977",
            "1978","1979","1980","1981","1982","1983")
tperiods <- length(time_id)
tperiods
```

```r
#calc Kronecker product using sparse matrix with Diagonal(tperiods)
I.T <- Diagonal(tperiods)

bigWmat <- kronecker(I.T,w1)
str(bigWmat)

#convert back to list weights object
bigW <- mat2listw(bigWmat,style="W")
summary(bigW)
```

```r
# Spatial Dependence Tests                                                                                           "RLMerr", "RLMlag", "SARMA"))
ctest1<- lm.LMtests(lm(mrate ~ legal, data = death1),bigW,test="all")
ctest1

# Spatial Dependence Tests                                                                                           "RLMerr", "RLMlag", "SARMA"))
ctest2<- lm.LMtests(lm(mrate ~ legal+state+year, data = death1),bigW,test="all")
ctest2
```

```r
# Spatial Dependence Tests                                                                                           "RLMerr", "RLMlag", "SARMA"))
ctest2<- lm.LMtests(lm(mrate ~ legal+state+year, data = death2),bigW,test="all")
ctest2
```
